```python
import requests
import bz2
import pandas as pd
from io import StringIO


url = "https://irsa.ipac.caltech.edu/data/download/wise-allwise/wise-allwise-cat-part01.bz2"
response = requests.get(url, stream=True)

decompressor = bz2.BZ2Decompressor()
buffer = b""
max_lines = 500000
count = 0
text_lines = []

for chunk in response.iter_content(chunk_size=1024 * 1024):  # 1 MB chunks
    buffer += decompressor.decompress(chunk)
    lines = buffer.split(b"\n")
    buffer = lines.pop()  # save last partial line for next chunk

    for line in lines:
        text_lines.append(line.decode('utf-8'))
        count += 1
        if count >= max_lines:
            break
    if count >= max_lines:
        break

# Join the lines into a single string with newline separators
text_data = "\n".join(text_lines)

# Now create a StringIO object so pandas can read it like a file
data_io = StringIO(text_data)

# Read with pandas using pipe separator
df = pd.read_csv(data_io, sep="|", header=None,
                 usecols=[1, 2, 16, 20, 24, 28],
                 names=['ra', 'dec', 'W1', 'W2', 'W3', 'W4'])

'''
0: name, 1: ra, 2: dec, 3: M1, 4: M2, 5: M3, 6: M4
'''

df_filter = df[(df["W4"]<9) | (df["W3"]<9)]
```

### **Astroquery**

**Catalog**: https://irsa.ipac.caltech.edu/cgi-bin/Gator/nph-dd?mode=html&catalog=allwise_p3as_psd

**Documentation**:
- Header reference: https://wise2.ipac.caltech.edu/docs/release/allwise/expsup/sec2_3a.html
- Young stellar object classification paper: https://ui.adsabs.harvard.edu/abs/2014ApJ...791..131K/abstract
- Astroquery: https://astroquery.readthedocs.io/en/latest/ipac/irsa/irsa.html
- Query: https://irsa.ipac.caltech.edu/applications/Gator/GatorAid/irsa/catsearch.html

```python
from astroquery.ipac.irsa import Irsa

query = "SELECT TOP 10 ra,dec,w1mpro,w2mpro FROM allwise_p3as_psd" 
results = Irsa.query_tap(query=query).to_qtable()
results
```

### **Graphing**

#### Scatter Plot - spatial distrobution

```python
import pandas as pd
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u

# Assuming your dataframe is already loaded and looks like this:
# df.columns = ['ID', 'RA', 'Dec', 'W1', 'W2', 'W3', 'W4']

# Convert to SkyCoord object (optional here, but useful if you want to do more with coordinates)
coords = SkyCoord(ra=df_filter['ra'].values * u.deg, dec=df_filter['dec'].values * u.deg, frame='icrs')

# Zoomed-in scatter plot of RA vs Dec near South Celestial Pole
plt.figure(figsize=(6, 6))
plt.scatter(df_filter['ra'], df_filter['dec'], s=10, color='red', alpha=0.7)
plt.xlabel('Right Ascension (deg)')
plt.ylabel('Declination (deg)')
plt.title('WISE Sources Near South Celestial Pole')
plt.grid(True)
plt.gca().invert_xaxis()  # optional, matches sky convention where RA increases right to left
plt.show()
```

#### Color plot - protostars

```python
import matplotlib.pyplot as plt

df_filter['W3-W4'] = df_filter['W3'] - df_filter['W4']

plt.figure(figsize=(7,5))
plt.scatter(df_filter['W3-W4'], df_filter['W4'], s=1, alpha=0.5, color='purple')
plt.xlabel('W3 - W4 (mag)')
plt.ylabel('W4 Magnitude')
plt.title('WISE Color-Magnitude Diagram')
plt.gca().invert_yaxis()  # Magnitude scale is inverted
plt.grid(True)
plt.show()
```
