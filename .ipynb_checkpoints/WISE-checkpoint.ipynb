#### **Definitions**

W1, W2,... : the magnitude of the frequency of light being detected > mag

W1: 3.4μm | W2: 4.6μm | W3: 12μm | W4: 22μm | ra: deg | dec: deg

mag = -2.5 * log10(flux) + constant

Flux is W/m^2/Hz -> energy received per square meter per unit time in a certain frequency (ie, 3.4μm).

```python
import requests
import bz2
import pandas as pd
from io import StringIO


url = "https://irsa.ipac.caltech.edu/data/download/wise-allwise/wise-allwise-cat-part01.bz2"
response = requests.get(url, stream=True)

decompressor = bz2.BZ2Decompressor()
buffer = b""
max_lines = 500000
count = 0
text_lines = []

for chunk in response.iter_content(chunk_size=1024 * 1024):  # 1 MB chunks
    buffer += decompressor.decompress(chunk)
    lines = buffer.split(b"\n")
    buffer = lines.pop()  # save last partial line for next chunk

    for line in lines:
        text_lines.append(line.decode('utf-8'))
        count += 1
        if count >= max_lines:
            break
    if count >= max_lines:
        break

# Join the lines into a single string with newline separators
text_data = "\n".join(text_lines)

# Now create a StringIO object so pandas can read it like a file
data_io = StringIO(text_data)

# Read with pandas using pipe separator
df = pd.read_csv(data_io, sep="|", header=None,
                 usecols=[1, 2, 24, 28],
                 names=['ra', 'dec', 'W3', 'W4'])

'''
1: ra, 2: dec, 24: W3, 28: W4
J200 Epoch
'''

df_filter = df[
    (df["W3"] - df["W4"] > 1.0) &
    (df["W4"] < 12) &
    (df["W3"] < 13)
]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u

# Assuming your dataframe is already loaded and looks like this:
# df.columns = ['ID', 'RA', 'Dec', 'W1', 'W2', 'W3', 'W4']

# Convert to SkyCoord object (optional here, but useful if you want to do more with coordinates)
coords = SkyCoord(ra=df_filter['ra'].values * u.deg, dec=df_filter['dec'].values * u.deg, frame='icrs')

# Zoomed-in scatter plot of RA vs Dec near South Celestial Pole
plt.figure(figsize=(6, 6))
plt.scatter(df_filter['ra'], df_filter['dec'], s=10, color='red', alpha=0.7)
plt.xlabel('Right Ascension (deg)')
plt.ylabel('Declination (deg)')
plt.title('WISE Sources Near South Celestial Pole')
plt.grid(True)
plt.gca().invert_xaxis()  # optional, matches sky convention where RA increases right to left
plt.show()
```

### Emission Range Plot


$$
\begin{align}
W3 = -2.5\log_{10}(\text{Flux}_{1}) \text{   |   } W4 = -2.5\log_{10}(\text{Flux}_{2}) \\ \\
\Delta \text{Magnitude} = W3 - W4 \\ \\
\Delta \text{Magnitude} = -2.5\log_{10}(\text{Flux}_{1}) + 2.5 \log_{10}(\text{flux}_{2}) \\ \\
        = -2.5 \log_{10}\left( \frac{\text{Flux}_{1}}{\text{Flux}_{2}} \right)
\end{align}
$$

#### We then compare this value to W4 in a plot. 

#### Longer wavelength -> redder

#### Since protostars and young stars mainly radiate in the far right infrared wavelength (due to the massive amounts of warm dust surrounding the star), we search for sources with a value of $\Delta \text{Magnitude} > 0$ indicating higher emissions in the 22μm range. 

```python
import matplotlib.pyplot as plt

# Compute color index
df_filter['W3-W4'] = df_filter['W3'] - df_filter['W4']

plt.figure(figsize=(7, 5))
plt.scatter(df_filter['W3-W4'], df_filter['W4'], s=2, color='red', alpha=0.5)

plt.xlabel('W3 - W4 (mag)')
plt.ylabel('W4 Magnitude')
plt.title('WISE Color–Magnitude Diagram')
plt.gca().invert_yaxis()  # Brighter at  top
plt.grid(True)
plt.show()

outliers = df_filter[df_filter['W3-W4'] > 5]
print(outliers)
```

#### Since mag is inversely related to the brightness, a low or negative magnitude indicates an incredible high brightness, and a high magnitude indicates a low brightness. 

#### A large $\Delta\text{Magnitude}$ value means that W3≫W4, meaning W3 is much fainter than W4, indicating high 22micrometer emissions.
